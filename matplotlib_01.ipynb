## Matplotlib
- python에서의 데이터 시각화 패키지

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

### 데이터 생성

```python
x = np.linspace(-5, 5, 11)
y = x ** 3 - x ** 2 - 10 * x
data = np.random.randint(10, size=10)
```

### 그래프 출력

```python
plt.scatter(x, y)
plt.show()
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(data)
plt.show()
```

- ### 그래프 크기 조절

```python
plt.figure(figsize=(10, 3)) # (가로, 세로)
plt.plot(data)
plt.show()
```

- ### x, y축 및 타이틀

```python
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('X-Y relation')
plt.plot(x, y)
plt.show()
```

- ### grid

```python
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('X-Y relation')
plt.grid(True)
plt.plot(x, y)
plt.show()
```

- ### x, y축 범위 지정

```python
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('X-Y relation')
plt.grid(True)

plt.xlim(-2, 10)
plt.ylim(-50, 100)

plt.plot(x, y)
plt.show()
```

- ### x, y축 tick 설정

```python
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('X-Y relation')
plt.grid(True)
plt.xlim(-2, 10)
plt.ylim(-50, 100)

plt.xticks(range(1, 5))
plt.yticks([0, 30])

plt.plot(x, y)
plt.show()
```

### 스타일 설정

- color
    - b : blue 
    - g : green
    - r : red
    - c : cyan 
    - m : magenta
    - y : yellow 
    - k : black 
    - w : white 

- marker

    - . : point
    - , : pixel
    - o : circle 
    - v : triangle_down
    - ^ : triangle_up 
    - < : triangle_left
    - \> : triangle_right 
    - 1 : tri_down
    - 2 : tri_up 
    - 3 : tri_left 
    - 4 : tri_right
    - s : square 
    - p : pentagon
    - \* : star 
    - h : hexagon1 
    - H : hexagon2 
    - \+ : plus 
    - x : x 
    - D : diamond 
    - d : thin_diamond

- line style

    - \- : solid 
    - \-- : dashed 
    - \-. : dash-dot 
    - : : dotted 
    

- 커스텀 스타일

    - c : 선 색상
    - lw : 선 굵기
    - ls : 선 스타일
    - marker : 마커 종류
    - ms : 마커 크기
    - mec : 마커 선 색깔
    - mew : 마커 선 굵기
    - mfc : 마커 내부 색깔


```python
plt.plot(x, y, 'g*')
plt.show()
```

```python
plt.plot(data, c="m", ls="-.", marker='h', lw=3, ms="12", mec="b", mew="3", mfc="y")
plt.show()
```

### 여러 그래프 출력하기

```python
datas = [np.random.randint(5, size=5) for _ in range(4)]
```

- #### plot 함수 한번 사용

```python
x = range(5)
plt.plot(x, datas[0], "ro--", x, datas[1], "b*-.", x, datas[2], "g^-", x, datas[3], "y<:")
plt.show()
```

- #### plot 함수 여러번 사용

- legend 범례

    - 위치 : 0 ~ 10
    - 0 : best 
    - 1 : ur
    - 2 : ul 
    - 3 : ll 
    - 4 : lr 
    - 5 : r 
    - 6 : cl 
    - 7 : cr 
    - 8 : lc 
    - 9 : uc 
    - 10 : center

```python
plt.plot(x, datas[0], label="data0")
plt.plot(x, datas[1], label="data1")
plt.plot(x, datas[2], label="data2")
plt.plot(x, datas[3], label="data3")
plt.legend(loc=0)
plt.show()
```

- #### subplot 사용

```python
for idx in range(4):
    plt.subplot(221 + idx)
    plt.plot(datas[idx])
```
