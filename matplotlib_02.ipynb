## Matplotlib
- 여러가지 차트 사용
- seaborn
- 데이터 프레임으로 데이터를 받아서 그래프를 그리기

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

### bar plot

```python
data = np.random.randint(10, size=7)

plt.bar(range(7), data, width=0.4)
plt.show()
```

```python
plt.bar(range(7), data, yerr=np.random.randint(1, 3, size=7), alpha=0.6)
plt.show()
```

```python
plt.barh(range(7), data, xerr=np.random.randint(1, 3, size=7), alpha=0.3)
plt.show()
```

```python
data1 = np.random.randint(1, 5, size=5)
data2 = np.random.randint(5, 10, size=5)
plt.bar(range(5), data2)
plt.bar(range(5), data1)
plt.show()
```

### pieplot

```python
plt.pie(data, labels=list("ABCDEFG"), colors=[
        "gold", "red", "blue", "lightgreen", 'brown', 'lightblue', "green"], autopct="%1.2f%%")
plt.axis("equal")
plt.show()
```

### scatter plot

```python
 # 균등분포
x = np.random.rand(200)
y = np.random.rand(200)
plt.scatter(x, y)
plt.show()
```

```python
# 정규분포
x = np.random.randn(200)
y = np.random.randn(200)
plt.scatter(x, y)
plt.show()
```

```python
# 균등분포, 정수형
x = np.random.randint(100, size=200)
y = np.random.randint(100, size=200)
plt.scatter(x, y)
plt.show()
```

```python
x = np.random.randint(10, size=20)
y = np.random.randint(10, size=20)
colors = np.random.rand(20)
area = np.pi * np.arange(10, 50, 2) ** 2
plt.scatter(x, y, c=colors, s=area, alpha=0.5)
plt.show()
```

### histogram

```python
y = np.random.randn(200)
plt.hist(y, bins=10, density=True)
plt.show()
```

### pairplot

```python
iris = sns.load_dataset("iris")
iris.tail(2)
```

```python
sns.lmplot("sepal_length", "sepal_width", data=iris, hue="species", fit_reg=False)
plt.show()
```

```python
sns.pairplot(iris, hue="species")
plt.show()
```

### heatmap

```python
flights = sns.load_dataset("flights")
flights_pivot = flights.pivot("month", "year", "passengers")
flights_pivot.tail()
```

```python
sns.heatmap(flights_pivot, cmap="YlGnBu", annot=True, fmt="d")
plt.show()
```
